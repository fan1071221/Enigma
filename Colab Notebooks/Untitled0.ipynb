```python
#0.553766806
import torch
import os
# Model
path = './res'  # 保存的路径
if not os.path.exists(path):
  os.makedirs(path)
text_name = 'res.txt'
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/drive/MyDrive/yolov5/runs/train/exp8/weights/best.pt',autoshape=False)
model.conf = 0.55  # confidence threshold (0-1)
model.iou = 0.45  # NMS IoU threshold (0-1)
model.classes = None  # (optional list) filter by class, i.e. = [0, 15, 16] for persons, cats and dogs
for i in range(3001,5501):
  imgs = '/content/drive/MyDrive/PrivateTestDataset/PrivateTestDataset/img/'+'img_'+str(i)+'.jpg' 
  #imgs = [dir + f for f in 'img_'+str(i)+'.jpg']  # batched list of images

  # Inference
  results = model(imgs)
  #results = model(imgs)  # includes NMS
  # Results
  results.print()  
  results.save()  # or .show()

  # Data
  #print(results.xyxy[0])  # print img1 predictions (pixels)
  res = results.xyxy[0].numpy()
  #print(results.xyxy[0].numpy())
  f_txt= open(os.path.join(path, text_name), 'a')
  count = i
  for i in range(res.shape[0]):
  
    x_min = res[i][0]
    y_min = res[i][1]
    x_max = res[i][2]
    y_max = res[i][3]
    conf = res[i][4]
    f_txt = open('./res/' + text_name,'a')
    f_txt.write("%d ,%d ,%d ,%d ,%d ,%d ,%d ,%d ,%d ,%f\n" % (count,round(x_min),round(y_min),round(x_max),round(y_min),round(x_max),round(y_max),round(x_min),round(y_max),conf))

f_txt.close()
```
